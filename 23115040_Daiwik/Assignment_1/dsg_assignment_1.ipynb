QUESTION 1(A)

(1) a) $X \sim \operatorname{Bin}(n, p)$
$$
P(X=x)={ }^n C_x p^x(1-p)^{n-x}
$$

To prove: $\left[\lim _{(n, p) \rightarrow(\infty, 0)} P(X=x)\right] \sim$ Poisson $(n p)$
Let $L=\lim _{(n, p) \rightarrow(\infty, 0)} p(X=x)$
$$
\begin{aligned}
& \Rightarrow L=\lim _{(n, p) \rightarrow(\infty, 0)} \frac{n!}{(n-x)!\cdot x!} p^x(1-p)^{n-x} \\
& =\lim _{(n, p) \rightarrow(\infty, 0)} \frac{(n p)^x(1-p)^n}{x!}\left[\frac{n!(1-p)^{-x}}{[(n-x)!n x}\right]
\end{aligned}
$$

Now, $(1-p)^{+n} \approx e^{-n p}$,
$$
\begin{aligned}
\cdot(1-p)^x & \approx 1^x=1[x \text { is finite }] \\
\frac{n!}{(n-x)! nx} & =\frac{n(n-1) \ldots(n-x+1)}{n^x} \\
& =\left(\frac{n}{n}\right)\left(1-\frac{1}{n}\right) \ldots\left(1-\frac{x-1}{n}\right) \\
& =(1)(1) \ldots(1) \rightarrow 1
\end{aligned}
$$

Hence,
$$
L=\frac{(n p)^x e^{-n p}}{x!} \sim \operatorname{Poisson}(n p)
$$

QUESTION 1(B)

```python
n=20
p=0.23

x = np.arange(1, n+1)
y_bin=np.array([])
for i in range(n):
    y=(np.math.comb(n,x[i]))*(p**x[i])*((1-p)**(n-x[i]))
    y_bin=np.append(y_bin,y)
```

```python
plt.bar(x, y_bin, align='center', alpha=0.75)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import poisson

poisson_dist = poisson(mu=n*p)

y_pois=poisson_dist.pmf(x)
```

```python
y_pois
```

```python
plt.bar(x, y_pois, align='center', alpha=0.75)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Poisson Distribution (lamda={4.6})')
plt.grid(True)
plt.show()
```

```python
#PLOTTING BOTH THE GRAPHS TOGETHER FOR COMPARISION

plt.bar(x, y_pois, align='center', alpha=0.75)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')

plt.grid(True)

plt.bar(x, y_bin, align='center', alpha=0.75)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')

plt.grid(True)

plt.title(' Binomial:Orange,  Poisson:Blue ')

plt.show()



```

```python
n=300000
p=10**(-6)

x_values = np.arange(1, 10)
y_bin_2=np.array([(np.math.comb(n,x))*(p**x)*((1-p)**(n-x)) for x in x_values])
```

```python
plt.bar(x_values, y_bin_2, align='center', alpha=1)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.grid(True)
plt.show()
```

```python
poisson_dist = poisson(mu=n*p)

y_pois_2=poisson_dist.pmf(x_values)
```

```python
plt.bar(x_values, y_pois_2, align='center', alpha=1)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Poisson Distribution (lambda={0.3})')
plt.grid(True)
plt.show()
```

In the first case where n was 20 and p was 0.23 , we observed that there were considerable deviations between the binomial and the poisson plots.

But in the 2nd case , where n tended to infinity and p tended to 0, there is hardly any difference between the two plots, hence justifying the result from 1(a).

QUESTION 6


Rejecting or Accepting the null hypothesis would depend on the chosen level of significance(alpha).


$$
\begin{aligned}
& H_0: p=0.5 \\
& H_1: p \neq 0.5
\end{aligned}
$$

If the chosen level of significance is $\alpha$.
then the critical region is:
$\left(-\infty, \bar{X}-z_{\alpha / 2} \frac{\sigma}{\sqrt{n}}\right) \cup\left(\bar{X}+z_{\alpha / 2} \frac{\sigma}{\sqrt{n}}, \infty\right)$

if the value of p proposed in the null hypothesis lies in the critical region, then we can reject the null hypothesis, otherwise, we fail to reject the null hypothesis

--->The null hypothesis is often chosen based on practical considerations or historical data, and 

p=0.5 is a common choice when there is no prior reason to expect a different value.
