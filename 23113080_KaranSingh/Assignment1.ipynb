QUESTION 1.\\
Let P(X=k) &= \binom{n}{k}p^k(1-p)^{n-k} \\

\text{For } p &\rightarrow \infty \text{ and } n \rightarrow 0 \quad \lambda=np\\

\lim_{n \to \infty} P(X=x) &= \lim_{n \to \infty} \binom{n}{k}p^k(1-p)^{n-k} \\

&= \lim_{n \to \infty} \frac{n!}{(n-k)!k!} p^k(1-p)^{n-k} \\

&= \lim_{n \to \infty} \frac{n!}{(n-k)!k!} (\frac{\lambda}{n})^k (1-\frac{\lambda}{n})^{n-k} \\

&= \frac{\lambda^k}{k!}\lim_{n \to \infty} \frac{n!}{(n-k)} (\frac{1}{n})^k (1-\frac{\lambda}{n})^n\\

&= \frac{\lambda^k}{k!}\lim_{n \to \infty} (n)(n-1)\dots(n-k+1) (\frac{1}{n})^k (1-\frac{\lambda}{n})^n\\

&= \frac{\lambda^k}{k!}\lim_{n \to \infty} (1)(1-\frac{1}{n})\dots(1-\frac{k}{n}+\frac{1}{n})(1-\frac{\lambda}{n})^n\\

&= \frac{\lambda^k}{k!}\lim_{n \to \infty} (1-\frac{\lambda}{n})^n\\

&= \frac{\lambda^k}{k!}e^{\lim_{n \to \infty}(1-\frac{\lambda}{n} -1)n}\\

&= \frac{\lambda^k}{k!}e^{-\lambda}\\

\end{align*}

```python
import matplotlib.pyplot as plt 
import numpy as np  
import scipy.special as sp
import scipy.stats as stats

n=20
p=0.23
lambda_parameter =n*p
factorial_n=sp.factorial(n)


k=np.linspace(0,n,n+1)
y_binomial =sp.comb(n,k)*pow(p,n)*pow((1-p),n-k)
y_poisson = ((lambda_parameter)**k)*np.e**(-lambda_parameter)/sp.factorial(k)
#y = stats.binom.pmf(k, n, p,5)
fig, axs =plt.subplots(2)
axs[0].plot(k,y_binomial,label="binomial",color="red")  
axs[0].set_title("Binomial")
axs[1].plot(k,y_poisson,label="poisson",color="blue")
axs[1].set_title("Poisson")
plt.show()
```
