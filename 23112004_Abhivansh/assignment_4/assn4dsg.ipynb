```python
import numpy as np
from sklearn.linear_model import LinearRegression

x= 2* np.random.rand(500,1)
y= 4+3*x + np.random.randn(500,1)
x.mean(),y.mean()
```

By calculations, the optimal value of theta(the parameters) is:
$$
\theta = (X^T X)^{-1} X^T y
$$

```python
x_b= np.c_[np.ones((500, 1)),x]
theta_optimal=np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)

y_p=x_b.dot(theta_optimal)
mse=np.mean((y_p-y)**2)
print("MSE:", mse)

print(f'Intercept:{theta_optimal[0]}')
print(f'Coefficients:{theta_optimal[1:]}')
```

-**Choice of Sigma ($\sigma$)**\
$\sigma$ here acted as a noise. Fixing sigma as Identity matrix makes the variance constant; but on the other hand, if sigma is variable (hyperparameter), then the model's flexibility tends to improve assigning different variances justifying the diversity of dataset.It allows the model to account for varying levels of uncertainty in predictions, potentially improving the fit and predictive performance.

-In Logistic Regression, the hypotheses $h_\theta (x)$ is taken to be:
$$
h_\theta (x)= g(\theta^T x)=\frac{1}{1+ e^{-\theta^T x}}
$$
where this function is termed as a "Sigmoid" function which ranges between open interval of (0,1). Logistic Regression is generally used for binary classification problems.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X= np.random.normal(0,1,200)
threshold=0.5
labels=(X>0.5).astype(int)
X+=np.random.normal(0,0.2,200)

plt.scatter(X[labels==0],np.zeros(sum(labels==0)),color="red",label="class 0")
plt.scatter(X[labels==1],np.ones(sum(labels==1)),color="green",label="class 1")
plt.axvline(x=threshold,color="blue",label="decision boundary")
plt.xlabel('Feature')
plt.ylabel("Class Label")
plt.legend()
plt.title('toy dataset')
plt.show()
```

Newton method:
$$
\theta^{t+1}=\theta^t - [H(\theta^t)]^{-1} \nabla L(\theta^t)
$$

```python
def sigmoid(z):
    return 1/(1+(np.e)**(-z))

def log_likelihood(X,y,ini_weights):
    m=len(y)
    pred=sigmoid(ini_weights.dot(X))
    return (np.sum(y*np.log(pred) + (1-y)*np.log(1-pred)))

def gradient_descent(X,y,weights,lr,iterations):
    m=len(y)
    for i in range(iterations):
        predictions = sigmoid(np.dot(X, weights))
        weights=weights + (lr*(X.T.dot(y)-X.T.dot(predictions)))/m
    return weights

lr=0.01
iterations=1000
X=np.random.normal(0,1,200).reshape(-1,1)
y=(X>0.5).astype(int).ravel()
X=np.hstack((np.ones((200,1)),X))
ini_weights=np.zeros(2)
final_weights=gradient_descent(X,y,ini_weights,lr,iterations)
predictions=sigmoid(X.dot(final_weights))
label_prediction=(predictions>0.5).astype(int)

print("Final weights:", final_weights)
print("Predictions:", label_prediction)
```

Fisher Scoring parameter update:
$$
\theta^{t+1}=\theta^t + s[I(\theta^{t})]^{-1} \nabla L(^{t})
$$

```python
def score_f(X,y,beta):
    p=sigmoid(X @ beta)
    return X.T @ (y-p)
def fisher_inf(X,beta):
    p=sigmoid(X @ beta)
    W = np.diag(p*(1-p))
    return X.T @ W @ X
def fisher_scoring(X,y,iterations,tolerance):
    tolerance=1e-6
    n,k=X.shape
    beta=np.zeros(k)
    for i in range(iterations):
        score=score_f(X,y,beta)
        fisher_info=fisher_inf(X,beta)
        update=np.linalg.inv(fisher_info) @ score
        beta +=update
        if np.linalg.norm(update)<tolerance:
            print(f'Converged after {i+1} iterations.')
            break
    return beta

true_beta=np.array([-0.5,1.0])
y=np.random.binomial(1, sigmoid(X@true_beta))

# Fit the model using Fisher Scoring
beta_hat=fisher_scoring(X,y,100,1e-6)
print("Estimated coefficients:", beta_hat)
```

Complexity Analysis:\
    -For NR method, the complexity is $O(nk)$ for Gradient computation, $O(nk^2)$ for Hessian computation and $O(k^3)$ for $k$ x $k$ matrix inversion.\
    -For FS method, the complexity is the same as mentioned for the NR.\
    -The NR method may require fewer iterations to converge because it uses the exact Hessian matrix, which contains more precise curvature information. However, this can also make it more sensitive to initial values and cause it to be less stable.

### *Exponential Family & GLMs*

>Ans. $Q-1$

Given : 
$$
p(y;\phi)= (1- \phi)^{1-y} \phi , y=1,2,3...
$$
We say that a class of distributions is in the exponential family if it can be written in the form :
$$
p(y;\eta)= b(y) e^{(\eta^T T(y) - a(\eta))}
$$
where $\eta$ is the cannonical parameter, $T(y)$ is the sufficient statistic and $a(\eta)$ is the log-partition function. And, $e^{a(\eta)}$ is the normalization constant.

**Part $a$.**
$$
p(y;\phi)= \phi (1-\phi)^{1-y}\\
p(y;\phi)=e^{(log(\phi) +(y-1)log(1- \phi))}\\
= e^{y(log(1-\phi))} e^{-log(1-\phi)} = \phi e^{(y(log(1-\phi)) - log(\frac{1-\phi}{\phi}))}
$$
Now comparing with the original equation, we can clearly observe the following:\
$ b(y)=1$ ,
$T(y)=y$ ,
$\eta = log(1-\phi)$ , 
$a(\eta)= -log(\frac{1-\phi}{\phi})$

**Part $b$.**

Using the three assumptions about the conditional distribution of $y$ (the *response variable*) given $x$, we can say that :
$$
h_\theta(x)= E[y/x;\theta]\\
= \mu = \frac{1}{1- e^{-\eta}}\\
= \frac{1}{1-e^{-\theta^T x}}
$$

**Part $c$.**

If we use the GLM modelling for geometric response, then the log likelihood can we written as : 
$$
\eta = \theta^T x
$$
$$
\begin{aligned} \ell_i(\theta) & =\log \left[\exp \left(\theta^T x^{(i)} \cdot y^{(i)}-\log \left(\frac{e^{\theta^T x^{(i)}}}{1-e^{\theta^T x^{(i)}}}\right)\right)\right] \\ &  =\theta^T x^{(i)} \cdot y^{(i)}+\log \left(e^{-\theta^T x^{(i)}}-1\right) \\ 
\frac{\partial}{\partial \theta_j} \ell_i(\theta) & =x_j^{(i)} y^{(i)}+\frac{e^{-\theta^T x^{(i)}}}{e^{-\theta^T x^{(i)}}-1}\left(-x_j^{(i)}\right) \\ &  =\left(y^{(i)}-\frac{1}{1-e^{\theta^T x^{(i)}}}\right) x_j^{(i)} .\end{aligned}
$$
So, Stochastic Gradient Ascent updation can be written as:
$$
\theta_{j+1}:= \theta{j} + \alpha \left(y^{(i)}-\frac{1}{1-e^{\theta^T x^{(i)}}}\right) x_j^{(i)} 
$$

Considering Gaussian Distribution for this problem. So the probability distribution function for gaussian dist. is :
$$
p(y;\mu,\sigma)= \frac{1}{\sqrt {2 \pi}} e^{-\frac{(y-\mu)^2}{2 \sigma^2}}
$$
The value of $\sigma^2$ has no effect on our final choice of $\theta$ and $h_{\theta}(x)$. Thus, we can choose an arbitrary value for $\sigma^2$ without changing anything. To simplify the derivation below, lets set $\sigma = 1 $. So now, 
$$
p(y;\mu,\sigma)= \frac{1}{\sqrt {2 \pi}} e^{-\frac{(y-\mu)^2}{2}} \\
= \frac{1}{\sqrt {2\pi}} e^{-\frac{y^2}{2}} e^{(\mu y - \frac{\mu^2}{2})}
$$
So, from here, we can conclude that, 
$b(y)=\frac{1}{\sqrt{2 \pi}} e^{-\frac{y^2}{2}}, T(y)=y, \eta=\mu, a(\eta)=\frac{\mu^2}{2}$

Hence it is a GLM member.

So, a general Linear Regression model would be ;
$$
y= w^T x + \epsilon
$$
where $w$ are the weights, $x$ is the input vector, and $\epsilon$ is the noise (normally distributed). Our objective is to find the weight $w$. So, first of all, we need to define our loss function, and then the gradient ascent.

```python
#exponential family distribution- Gaussian
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
n=100
X=3*np.random.rand(n,1)
y=4+3*X+np.random.randn(n,1)

#adding the bias term
X_b=np.c_[np.ones((n,1)), X]
weights=np.random.randn(2,1)

lr=0.01
iterations=1000

for iteration in range(iterations):
    gradients=2/n * X_b.T.dot(X_b.dot(weights)-y)
    weights-=lr*gradients

print("Estimated weights:", weights)

plt.plot(X,y,"b.")
plt.plot(X,X_b.dot(weights),"r-")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression with Gaussian Noise")
plt.show()
```
