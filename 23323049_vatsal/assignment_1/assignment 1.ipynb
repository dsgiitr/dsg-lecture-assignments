# 1.a) Deriving the Poisson Distribution from the Binomial Distribution

### The Binomial Distribution


$\ P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n-k} \$


### Conditions for the Poisson Distribution

The Poisson distribution arises as a limiting case of the binomial distribution under certain conditions:

1. The number of trials \$ n \$, is large.
2. The probability of success \$ p \$ is small.
3. The product \$ n \cdot p \$ remains constant.

Substitute \$ p = \frac{\lambda}{n} \$ into the binomial PMF:

\$ P(X = x) = \binom{n}{x} \cdot \left(\frac{\lambda}{n}\right)^x \cdot \left(1 - \frac{\lambda}{n}\right)^{n-x} \$

As \$ n \$ becomes large, we can rewrite the binomial coefficient and apply limits:

\$ P(X = x) = \lim_{n \to \infty} \frac{n!}{x!(n-x)!} \cdot \left(\frac{\lambda}{n}\right)^x \cdot \left(1 - \frac{\lambda}{n}\right)^{n-x} \$


\$ P(X = k) = \lim_{n \to \infty} \frac{n!}{x!(n-x)!}  \cdot \frac{\lambda^x}{n^x} \cdot \left(1 - \frac{\lambda}{n}\right)^{n} \cdot \left(1 - \frac{\lambda}{n}\right)^{-x} \$
As \$ n \to \infty \$:

- \$ \left(1 - \frac{\lambda}{n}\right)^{n} \to e^{-\lambda} \$
- \$ \left(1 - \frac{\lambda}{n}\right)^{-k} \to 1 \$
- \$ \frac{n!}{(n-x)!n^x} = \frac{(n) (n-1)...(n-x+1)}{n^x}\approx 1 \$ for large \$ n \$






\$ P(X = k) = \frac{e^{-\lambda} \cdot \lambda^k}{k!} \$

which is the PMF of the Poisson distribution with parameter \$ \lambda \$.

# 3. WLLN statement using chebychev's inequality
This proof uses the assumption of finite variance 
$Var(X_{i})=\sigma ^2$
The independence of the random variables implies no correlation between them, and we have that
$$Var( \bar X ) =Var(\frac{1}{n}(X_{1}+X_{2}+ ... +X_{n})=\frac{\sigma ^2}{n}$$
Similarly, we have sample mean = $\mu$

Using Chebyshev's inequality on 
$\bar X_{n}$ :
$P( |{\bar X_{n}-\mu}|\ge\epsilon)\le\frac{\sigma^2}{n\epsilon^2}$

$lim_{n->\infty}P( |{\bar X_{n}-\mu}|\ge \epsilon)=0$ because probability can not be negative

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


enrollment_number = '23323049'
digits = [int(d) for d in enrollment_number]
sum_digits = sum(digits)
probabilities = [d / sum_digits for d in digits]


print("Digit probabilities:")
s=0
m=0
for i, prob in enumerate(probabilities):
    s=s+i*prob
    m=m+i**2*prob
    print(f"P(X={i+1}) = {prob}")
print(f"mean={s}") 
print(f'std={np.sqrt(m-s**2)}')
means = []
for _ in range(10000):
    sample = np.random.choice(digits, size=10000)
    means.append(np.mean(sample))
exp_mu = np.mean(means)
exp_sigma = np.std(means)
print(f"Experimental mean (mu): {exp_mu}")
print(f"Experimental standard deviation (sigma): {exp_sigma}")
    
sns.histplot(means, kde=True)
plt.xlabel('Sample Means')
plt.ylabel('Frequency')
plt.title('Histogram of Sample Means (10,000 repetitions)')
plt.show()
```

the resultant graph can be approximated by a normal distribution which largelty conforms the central limit theorem.
