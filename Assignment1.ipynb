**Question 1a:**<br>
$n \to \infty$ <br>
$p \to 0$<br>
$P(X=x) = \binom{n}{x}p^x(1-p)^x$<br><br>
$P(X=x)=\frac{((n) (x) ... (n-x+1))}{x!}\frac{\lambda^x}{n^x}(1-p)^{n-x}$<br>
$\lim_{n\to\infty}\frac{1(1-\frac{1}{n})(1-\frac{2}{n})...(1-\frac{x-1}{n})}{x!}\lambda^x(1-\frac{\lambda}{n})^{n-x}$<br><br>
as $n\to\infty$ the pdf will reduce to $\lim_{n\to\infty}\frac{\lambda^x}{x!}\frac{(1-\frac{\lambda}{n})^n}{(1-\frac{\lambda}{n})^x}$<br>
given wkt $\lim_{n\to\infty}(1+\frac{1}{n})^n=e$ then $\lim_{n\to\infty}(1-\frac{\lambda}{n})^n=e^{-\lambda}$
<br><br>
using this we get the pdf of poisson distribution as $P(X=x)=\frac{\lambda^xe^{-\lambda}}{x!}$

```python
import numpy as np 
import matplotlib.pyplot as plt
import math

mu1 = 6 
mu2 = 4  
mu3 = 2 

def poisson_pmf(k, mu): 
    return np.exp(-mu) * (mu ** k) / math.factorial(k) 
def binomial_pmf(k,n,p):
    return (p**k)*(math.factorial(n)/(math.factorial(n-k)*math.factorial(k)))*((1-p)**(n-k))


x = np.arange(0, 10) 
pmf1 = np.array([poisson_pmf(k, mu1) for k in x]) 

pmf2 = np.array([poisson_pmf(k, mu2) for k in x]) 

pmf3 = np.array([poisson_pmf(k, mu3) for k in x]) 

plt.plot(x, pmf1, label='Poisson Distribution with $\lambda$=6', alpha=0.5) 
plt.plot(x, pmf2, label='Poisson Distribution with $\lambda$=4', alpha=0.5) 
plt.plot(x, pmf3, label='Poisson Distribution with $\lambda$=2', alpha=0.5) 
plt.xlabel('x') 
plt.ylabel('P(X=x)') 
plt.title('Poisson Distribution') 
plt.legend() 
plt.grid(True) 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5, 6, 7, 8])
P = np.array([2/24, 3/24, 1/24, 1/24, 5/24, 1/24, 2/24, 9/24])

means = []
for _ in range(10000):
    samples = np.random.choice(X, size=10000, p=P)
    mean = np.mean(samples)
    means.append(mean)

# Plot histogram
plt.hist(means, bins=30, edgecolor='black')
plt.title('Histogram of Means')
plt.xlabel('Mean')
plt.ylabel('Frequency')
plt.show()
```

```python
mean = 5.5
std_dev = 0.02483

x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 1000)
pdf = 1/(std_dev * np.sqrt(2 * np.pi)) * np.exp(-(x - mean)**2 / (2 * std_dev**2))

plt.plot(x, pdf)
plt.title("Normal Distribution with Mean = {} and Standard Deviation = {}".format(mean, std_dev))
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
